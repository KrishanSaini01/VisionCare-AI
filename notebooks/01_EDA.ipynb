```python
normal_path = os.path.join(train_path, "NORMAL")

images = [
    f for f in os.listdir(normal_path)
    if f.lower().endswith((".jpg", ".jpeg", ".png"))
]

image_path = os.path.join(normal_path, images[0])

print(image_path)
print(os.path.exists(image_path))

import cv2

img = cv2.imread(image_path)

print(img.shape)
```

```python
import matplotlib.pyplot as plt
import cv2
import os

plt.figure(figsize=(12,6))

images = os.listdir(normal_path)

for i in range(6):
    img = cv2.imread(os.path.join(normal_path, images[i]))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.subplot(2,3,i+1)
    plt.imshow(img)
    plt.title("NORMAL")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
pneumonia_path = os.path.join(train_path, "PNEUMONIA")

plt.figure(figsize=(12,6))

images = os.listdir(pneumonia_path)

for i in range(6):
    img = cv2.imread(os.path.join(pneumonia_path, images[i]))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.subplot(2,3,i+1)
    plt.imshow(img)
    plt.title("PNEUMONIA")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
corrupted = 0

for cls in ["NORMAL", "PNEUMONIA"]:
    folder = os.path.join(train_path, cls)

    for file in os.listdir(folder):
        img = cv2.imread(os.path.join(folder, file))

        if img is None:
            corrupted += 1

print("Corrupted Images:", corrupted)
```
